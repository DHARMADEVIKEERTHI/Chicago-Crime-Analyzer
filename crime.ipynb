```python
import pandas as pd
from sqlalchemy import create_engine

# Database connection
engine = create_engine("postgresql+psycopg2://{user}:{pw}@{host}/{db}".format(
    user="postgres", 
    pw="boot", 
    host="localhost", 
    db="Chicago"
))

# Read the CSV file
data = pd.read_csv('Cleaned Crime_Data.csv')

# Rename columns (if needed)
data.columns = [
    "ID",  # Adjust column names to match the table schema
    "Case_Number",
    "date_time",
    "Block",
    "IUCR",
    "Primary_Type",
    "Description",
    "Location_Description",
    "Arrest",
    "Domestic",
    "Beat",
    "District",
    "Ward",
    "Community_area",
    "FBI_Code",
    "X_Coordinate",
    "Y_Coordinate",
    "Year",
    "updated_on",
    "Latitude",
    "Longitude",
    "Location"
]

# Ensure data types match the database schema
data["date_time"] = pd.to_datetime(data["date_time"])
data["updated_on"] = pd.to_datetime(data["updated_on"])

# Insert data into the PostgreSQL table
data.to_sql('crime_data', con=engine, if_exists='append', index=False, chunksize=10000)

print("Data inserted successfully.")
```
